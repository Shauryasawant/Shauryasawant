```python
import io
import csv
from bs4 import BeautifulSoup
from google.colab import files

# Prompt user to upload HTML file
print("Please upload the HTML file:")
uploaded = files.upload()

# Ensure only one file is uploaded
html_filename = next(iter(uploaded))

# Read and parse the uploaded HTML file
with open(html_filename, 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

# Find all div elements with class "local"
local_divs = soup.find_all('div', class_='local')

# Prepare data for CSV
data = []
for div in local_divs:
    name_elem = div.find('p', class_='name tit05')
    address_elem = div.find('p', class_='add c-gray1 mgt24')
    if name_elem and address_elem:
        name = name_elem.text.strip()
        address = address_elem.text.strip()
        data.append([name, address])
    else:
        print("Missing required element in div:", div)

# Create CSV in memory
output = io.StringIO()
csv_writer = csv.writer(output)
csv_writer.writerow(['Name', 'Address'])  # Header
csv_writer.writerows(data)

# Save CSV file
csv_filename = 'mobis_locations.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as f:
    f.write(output.getvalue())
print(f"CSV file '{csv_filename}' saved successfully.")

# Download the CSV file
files.download(csv_filename)

# Print the extracted data to the console
print("\nExtracted Data:")
for row in data:
    print(f"Name: {row[0]}")
    print(f"Address: {row[1]}")
    print("-" * 50)
```
